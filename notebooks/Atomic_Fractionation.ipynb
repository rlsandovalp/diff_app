```python
import numpy as np
from scipy.special import erf
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
data_Bolai = pd.read_csv('../Source_data/Bolai_All.txt', sep = '\t')
```

```python
def model(Cr, D, t, Z):
    C = np.zeros(len(Z))
    t = t*1000000*365*86400
    for pos_z, z in enumerate(Z):
        C[pos_z] = (1-math.erf(z/(2*math.sqrt(D*t))))*(1-Cr)+Cr
    return C
```

### Main ideas behind: Prinzhofer and Pernaton 1997
- Fractionation of isotopologues of methane occurs given that they have different values of diffusion coefficient
- Fractionation is enhanced in environments with low Knudsen number
- Usually $\delta^{13}C_1$ values decrease as one approaches the surface
- Low values [-110, -60] of $\delta^{13}C_1$ (aka enriched with $C^{12}$) are associated with biogenic methane
- High values [-60, -20] of $\delta^{13}C_1$ (aka depleted of $C^{12}$) are associated with termogenic methane
- Fraccionation of methane can also produce gas enriched with $C^{12}$

Let us name $A$, $B$, $C$, and $D$ four absolute ammounts for which we can measure the relative ratios $A/B$ and $C/D$. If we assume that there are 2 end-members the ratios $X=A/B$ and $Y=C/D$ of the mixture can be formulated as:

$$
X = \frac{(1-\xi)A_1+\xi A_2}{(1-\xi)B_1+\xi B_2}
$$

$$
Y = \frac{(1-\xi)C_1+\xi C_2}{(1-\xi)D_1+\xi D_2}
$$

where $\xi$ is the proportion of end-member 2.

The equations can be manipulated, an one can write:
$$
XY(B_1D_2-B_2D_1)+X(B_2C_1-B_1C_2)+Y(A_2D_1-A_1D_2)+A_1C_2-A_2C_1=0
$$

Under the assumption that the absolute ammount B is the same absolute ammount D one has:
$$
X(B_2C_1-B_1C_2)+Y(A_2B_1-A_1B_2)+A_1C_2-A_2C_1=0
$$

$$
Y = -X\frac{(B_2C_1-B_1C_2)}{(A_2B_1-A_1B_2)}+\frac{(A_2C_1-A_1C_2)}{(A_2B_1-A_1B_2)}
$$

That is the equation of a line. Lets assume that $A = [C_2]$, $B = [C_1] \approx [C_1^{12}]$, and $C = [C_1^{13}]$.

Then Y varies linearly with X with slope $m = \frac{(B_2C_1-B_1C_2)}{(A_2B_1-A_1B_2)}$ and intersection point $b = \frac{(A_2C_1-A_1C_2)}{(A_2B_1-A_1B_2)}$.

Lets assume that we have a mixture between two end-members with $X=\frac{C_2}{C_1}$ varying between 0 and 0.3 and $Y=\frac{C_1^{13}}{C_1^{12}}$ varying between 0.0103058 and 0.0108099 (or -80 and -35 in terms of $\delta^{13}C_1$). 

```python
delta = np.linspace(-80,-35,100)
c2c1 = np.linspace(0,0.3,100)

with plt.xkcd():
    plt.plot(delta, c2c1, '.')
    plt.xlabel(r'$\delta^{13}C_1$')
    plt.ylabel(r'$C_2/C_1$')
    plt.show()
```

Now, lets compute the evolution of the concentration profile of $C_1$ and $C_2$ with the analytical solution of the transport equation

$$
C(z,t) = C_o\left[1-erf\left(\frac{z}{2\sqrt{Dt}}\right)\right]
$$

```python
betas = [0.9, 0.75, 0.65, 0.5]

with plt.xkcd():
    plt.plot(delta, c2c1, '--', label = 'Mixture')
    for beta in betas:
        D_C1_12 = 10
        alpha = 0.995
        D_C1_13 = D_C1_12*alpha
        D_C2 = D_C1_12*beta

        t = 30

        Co_C1_12 = 1
        Co_C1_13 = Co_C1_12*0.0108099
        Co_C2 = Co_C1_12*0.3

        z = np.linspace(0,100,101)

        C1_12 = Co_C1_12*(1-erf(z/(2*np.sqrt(D_C1_12*t))))
        C1_13 = Co_C1_13*(1-erf(z/(2*np.sqrt(D_C1_13*t))))
        C2 = Co_C2*(1-erf(z/(2*np.sqrt(D_C2*t))))

        delta_13 = (C1_13/C1_12-0.011202)*1000/0.011202
        C2C1_m = C2/C1_12

        plt.plot(delta_13, C2C1_m, label = 'beta ='+str(beta))
    plt.xlabel(r'$\delta^{13}C_1$')
    plt.xlim(-80,-30)
    plt.ylabel(r'$C_2/C_1$')
    plt.legend()
    plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize = (12,6))
ax[0].plot(C1_12, z)
ax[1].plot(C1_13, z)
ax[2].plot(C2, z)
plt.show()
```
